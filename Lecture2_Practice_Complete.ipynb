### 1.2 Understanding Pandas DataFrame

In this section, we focus on indexing and subsampling of a pandas dataframe. This helps with calculation and advanced visualization in upcoming sections.

One advantage of using Pandas DataFrame is to obtain columns (in pandas, a 'Series') easily through "[ ]" operator. 
- To obtain one variable, use: **DataFrame[var_name]**
- To obtain multiple variables, use: **DataFrame[var_list]**, where var_list = [var1, var2, ...]

```python
# Selecting one/multiple columns through [] operator
# var = ['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF']

# data['SalePrice']

# data['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF'] # wrong, not a list

var = ['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF']

data[var]

data[['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF']]
# 'SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF'
```

There are two choices for indexing in a DataFrame. 
- **Integer based indexing.** Works similar to numpy array. **Syntax: DataFrame.iloc[row_range, column_range]**
    - You can use a list of integers to select separated elements.
- **Label based indexing.** Extract rows and columns based on corresponding label (e.g., variable name). Syntax: **DataFrame.loc[row_range, column_range]**. 
    - For label indexing, the range is INCLUSIVE!

```python
data.head(10)
```

```python
# Select rows and columns based on integer indexing. 
# This works the same as 2d arrays.
data.iloc[:, 0:3]
```

```python
# Selecing several columns/rows based on label indexing 
# By default, row labels are just integers

data.loc[ : , 'Id':'MSZoning']
```

### 2.1 Visualization Basics
**Labels and Titles**

- Use **plt.xlabel("CONTENT")**, and **plt.ylabel("CONTENT")** to specify the labels. Use **plt.title("TITLE NAME")** to specify the titles.

- Use **plt.xlim(LEFT_LIMIT, RIGHT_LIMIT)** to set range for x axis. Change xlim to ylim for y axis.

- Use **plt.xticks(ticks, labels)** to specify ticks (i.e., what values to include) and **corresponding** labels (i.e., labels of the values). Both should be specified in a list. The length of ticks and labels must match.

- Use **plt.legend([list_of_legends])** to show legend

**Parameter Settings**

The specific parameters differ for different plots. But here are some common parameters.
- color: color of the lines/bars. 'r', 'g', 'b', 'k' for red, green, blue, and black, correspondingly.
- alpha: transparency level. 0<alpha<=1; lower value of alpha => more transparent.
- linestyle: the line style (e.g., solid/dash/...). Solid: -, Dashed: --, Dotted: ., Dash-dot: -.

### Section 2.3 Boxplot
Use syntax **plt.boxplot(x)** to create boxplot of x.

It is very tedious to draw a boxplot of different groups using pyplot. This case, use: **sns.boxplot(data, x, y...)**
- x: If one-dimensional, then represents the variables to plot. If two-dimensional (i.e., box plot of different groups), then (by default) represents the group indicator.
- y: Not available if one-dimensional. If two-dimensional, then (by default) the variable to plot.
- orient: = "v" if vertical (default), ="h" if horizontal
- Need to specify variables clearly (including dataframe)

For other specifications: https://seaborn.pydata.org/generated/seaborn.boxplot.html

### Section 4.2 Data Scaling and Encoding
Data scaling and encoding syntax are under Scikit-Learn package Preprocessing Module. Coding when using sklearn packages is longer and follows a standard workflow. We will elaborate the workflow in linear regression section. In the current section, we only provide the syntax and coding template.

For specific parameter settings, please check the sklearn manual. You can simple copy and paste the syntax in Google to find the official manual.

**Standard Scaler:** For standardization (i.e., applying standard scaler), use syntax:

**<center>sklearn.preprocessing.StandardScaler()</center>**

**Min-Max Scaler:** For min-max scaler, use syntax:

**<center>sklearn.preprocessing.MinMaxScaler()</center>**

**One-Hot Encoding:** 
For One-hot encoding, use syntax: 

**<center>sklearn.preprocessing.OneHotEncoder()</center>**

**Ordinal Encoding**
For ordinal encoding, use syntax:
**<center>sklearn.preprocessing.OrdinalEncoder()</center>**

```python
# Try Scaling: original unscaled features
var = ['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF']
X = df[var]
X.describe()
```

```python
# Standard scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()                    # define scaler
X_standardized = scaler.fit_transform(X)     # obtain transfered X. Returns a Numpy array

# The scaled data is: X_standardized
```

```python
# Now let's check the summary statistics of X_standardized
## (For simple summary, change 2D numpy array to pandas dataframe, and get summary statistics)
new_dataframe = pd.DataFrame(X_standardized, columns = var)
new_dataframe.describe()  # focus on mean and std
# X_standardized.mean(axis=0) # This provides the mean of each variable
```

```python
# MinMax Scaler
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()

X_minmax = scaler_minmax.fit_transform(X)

# The scaled data is: X_minmax
```

```python
# Similarly, check summary statistics
new_dataframe_minmax = pd.DataFrame(X_minmax, columns = var)
new_dataframe_minmax.describe() # focus on min and max
```

```python
# Let's first collect two categorical variables
catvar = ['SaleType', 'SaleCondition']
X_cat = df[catvar]   # X_cat: two categorical variables
df.SaleType.value_counts(), df.SaleCondition.value_counts()  # this shows the unique values under each categorical var
```

```python
X_cat
```

```python
# One-hot Encoder
from sklearn.preprocessing import OneHotEncoder
my_encoder = OneHotEncoder(handle_unknown='ignore', sparse = False) # handle_unknown: whether to report error when missings are found
                                                                    # sparse: set it to True improves efficiency. Currently set to False for presentation purpose
X_encoded = my_encoder.fit_transform(X_cat)
# Transfered data: X_encoded
```

```python
print(X_encoded.shape) # we have 15 binary features (recall we have 15 )
print(my_encoder.categories_) # This shows all the categories. The variables are generated based on the order of the specific array.
```
